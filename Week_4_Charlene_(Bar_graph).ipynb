```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
##

import pandas as pd


crime_data = pd.DataFrame({
    "Crime": ["Murder", "Rape", "Robbery", "Aggravated Assault"],
    "Rate_2022": [6.3, 40.0, 67.2, 273.0],   # per 100k people (FBI estimates)
    "Rate_2023": [6.3 * (1 - 0.116), 40.0 * (1 - 0.094), 67.2 * (1 - 0.003), 273.0 * (1 - 0.028)]
})

crime_data
```

```python
##Bar chart of 2023 crime rates

plt.figure(figsize=(10,6))

colors = sns.color_palette("viridis", len(crime_data))

sns.barplot(x="Crime", y="Rate_2023", data=crime_data, palette=colors)
plt.title("U.S. Estimated Crime Rates in 2023 (per 100,000 People)", fontsize=16)
plt.xlabel("Crime Type", fontsize=12)
plt.ylabel("Rate per 100k", fontsize=12)

for i, val in enumerate(crime_data["Rate_2023"]):
    plt.text(i, val + 2, f"{val:.1f}", ha='center', fontsize=10)

plt.show()
```

This code creates a small dataset using real FBI crime statistics to compare crime rates from 2022 to 2023.

The dataset includes four major violent crimes: Murder, Rape, Robbery, and Aggravated Assault.

The column Rate_2022 shows the estimated crime rate per 100,000 people in 2022, based on FBI national data.

The column Rate_2023 is calculated by applying the official percent decreases reported by the FBI to the 2022 rates.

```python
##Horizontal bar chart of % changes (20222023)


crime_data["Percent Change"] = ((crime_data["Rate_2023"] - crime_data["Rate_2022"]) /
                               crime_data["Rate_2022"]) * 100

plt.figure(figsize=(10,6))
sns.barplot(x="Percent Change", y="Crime", data=crime_data, palette="coolwarm")

plt.title("Percent Change in Crime Rates (2022  2023)", fontsize=16)
plt.xlabel("Percent Change (%)", fontsize=12)
plt.ylabel("Crime Type", fontsize=12)

for i, val in enumerate(crime_data["Percent Change"]):
    plt.text(val - 1 if val < 0 else val + 0.5, i, f"{val:.1f}%", va='center', fontsize=10)

plt.show()
```

All crime categories have negative percent values, which means crime decreased across the board in 2023.

Murder shows the largest decrease, with a drop of about 11.6%, making it the most significant change.

Rape also decreased noticeably, by around 9.4%, showing a meaningful reduction.

Aggravated assault had a smaller decline of about 2.8%, indicating a more gradual change.

Robbery changed the least, with less than a 1% decrease, meaning its rate stayed relatively stable.

The horizontal layout makes it easy to compare how much each crime changed, and the cool-to-warm color scale helps highlight which crimes saw larger decreases. Overall, this graph clearly shows that violent crime declined nationally from 2022 to 2023.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline

##Horizontal Bar Chart  Percent Change (2022  2023)


crime_data["Percent Change"] = ((crime_data["Rate_2023"] - crime_data["Rate_2022"]) /
                               crime_data["Rate_2022"]) * 100

plt.figure(figsize=(10,6))
sns.barplot(x="Percent Change", y="Crime", data=crime_data, hue="Crime", palette="coolwarm", legend=False)

plt.title("Percent Change in Crime Rates (2022  2023)", fontsize=16)
plt.xlabel("Percent Change (%)", fontsize=12)
plt.ylabel("Crime Type", fontsize=12)

for i, val in enumerate(crime_data["Percent Change"]):
    plt.text(val - 1 if val < 0 else val + 0.5, i, f"{val:.1f}%", va='center', fontsize=10)

plt.show()
```

Murder experienced the largest decrease, showing a drop of about 11.6%, which is a significant improvement.

Rape also declined noticeably, decreasing by around 9.4%.

Aggravated assault decreased slightly, showing a smaller change of about 2.8%.

Robbery had the smallest change, with less than a 1% decrease, meaning it stayed fairly stable compared to other crimes.

```python
##MultiSeries Grouped Bar Chart (Fake Company Data)

grouped_data = pd.DataFrame({
    "Company": ["A","A","A","B","B","B","C","C","C"],
    "Product": ["1","2","3","1","2","3","1","2","3"],
    "Sales": [20,35,30,25,40,45,30,25,50]
})

pivot = grouped_data.pivot(index="Company", columns="Product", values="Sales")

# Use a bright, vibrant palette
pivot.plot(
    kind="bar",
    figsize=(10,6),
    color=sns.color_palette("tab10", 3)  # bright & harmonized colors
)

plt.title("Company Sales by Product (Grouped Bar Chart)", fontsize=16, weight="bold")
plt.xlabel("Company", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.legend(title="Product")
plt.xticks(rotation=0)
plt.show()
```

This grouped bar chart shows sales for three products across three companies using bright, easy-to-distinguish colors.

Company A performs best with Product 2, while Products 1 and 3 have lower sales.

Company B shows strong sales overall, especially for Product 3, which is its top product.

Company C has the highest sales for Product 3 out of all companies, making it the strongest performer in the chart.

Across all companies, Product 3 generally has the highest sales, while Product 1 tends to have the lowest.

The bright, harmonized colors make it easy to compare products within each company and see differences across companies. Overall, the chart clearly highlights which products perform best and which companies lead in sales.
