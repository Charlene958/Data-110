### Step 2: Anscombe's Quartet Dataset

This dataset is known as **Anscombe's Quartet**, created by statistician Francis Anscombe to illustrate the importance of visualizing data. Despite having nearly identical statistical properties (e.g., mean, variance, correlation, and linear regression), each dataset tells a very different story when graphed.

- **x**: The independent variable, common across three datasets.
- **y1, y2, y3**: Three different dependent variables associated with the same `x` values.
- **x4, y4**: A special case where most of the `x` values are identical, with one outlier.

#### Anscombe's Quartet:

`

```python
# Anscombe's Quartet:
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
plt.scatter(x, y1)
```

```python
# Calculate the slope (m) and intercept (b) of the line using np.polyfit
m, b = np.polyfit(x, y1, 1)

# Create the regression line
regression_line = m * np.array(x) + b

# Plot the data points and regression line
plt.scatter(x, y1)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y1')
```

### Your Task:

Perform the same linear regression process for the following datasets: y2, y3, and y4.
Modify the code to calculate and plot the regression lines for each of these datasets.
Use distinct colors for each plot and appropriately label the axes (y2, y3, etc.).
Discuss any differences you observe when comparing the results across all datasets.


```python
# Your Code need to be here for x and y2

import numpy as np
import matplotlib.pyplot as plt

x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y2 = np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])

m, b = np.polyfit(x, y2, 1)

y_pred = m * x + b
ss_res = np.sum((y2 - y_pred) ** 2)
ss_tot = np.sum((y2 - np.mean(y2)) ** 2)
r2 = 1 - ss_res / ss_tot if ss_tot != 0 else float('nan')

print("Anscombe Dataset 2 (x vs y2) — Linear regression results")
print(f"  slope (m)     = {m:.4f}")
print(f"  intercept (b) = {b:.4f}")
print(f"  R^2           = {r2:.4f}")

plt.figure(figsize=(7,5))
plt.scatter(x, y2, color='tab:orange', edgecolor='k', label='data (y2)')
x_line = np.linspace(x.min()-1, x.max()+1, 100)
plt.plot(x_line, m * x_line + b, color='tab:red', linestyle='--', linewidth=2, label=f'fit: y={m:.2f}x+{b:.2f}')
plt.xlabel('x')
plt.ylabel('y2')
plt.title('Anscombe — Dataset 2: y2 vs x with regression line')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)

plt.annotate(f"m={m:.3f}\nb={b:.3f}\nR²={r2:.3f}", xy=(0.02, 0.95), xycoords='axes fraction',
             fontsize=10, verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", alpha=0.8))
plt.show()
```

Overall, these four datasets show why plotting data matters. At first glance their summary numbers (means, variances, regression slope) look almost the same, but the scatterplots tell very different stories.

How they differ visually: Dataset 1 looks like a normal linear relationship with some scatter around the line. Dataset 2 shows a curved pattern (not a straight line) even though the linear fit looks similar numerically. Dataset 3 has one point with a very high y that pulls things upward, and Dataset 4 has many points with the same x and one clear outlier at x = 19. So the shapes and point patterns are very different even though the summary stats match.

Effect of the outlier in x4,y4: The single outlier (x = 19) heavily pulls the least-squares line toward it, so the fitted line does not represent the majority of the points. That one point changes the slope and intercept a lot, showing how sensitive ordinary least squares is to influential observations.

Why visualize as well as compute stats: Summary statistics compress data into a few numbers and can hide important features like nonlinearity, clusters, outliers, or influential points. Plotting helps you see the real pattern and decide if a linear model is appropriate or if you need a different approach (robust regression, transform, or a nonlinear model)
