```python
## Visualization 1: Line Chart – ED Visits Over Time

df_visit_counts = df[df['Estimate Type'] == 'Visit count']

visits_by_year = df_visit_counts.groupby('Year')['Estimate'].sum().reset_index()

plt.figure(figsize=(11, 6))
plt.plot(
    visits_by_year['Year'],
    visits_by_year['Estimate'],
    marker='o',
    markersize=8,
    linewidth=3.5,
    color='#972eb9',
    label='Total ED Visits'
)

plt.fill_between(
    visits_by_year['Year'],
    visits_by_year['Estimate'],
    color='#b92eb5',
    alpha=0.35
)
plt.title(
    'Emergency Department Visits in the U.S. (2016–2020)',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Year', fontsize=13)
plt.ylabel('Number of ED Visits', fontsize=13)

plt.grid(alpha=0.3, linestyle='--')

plt.annotate(
    'Sharp decline in 2020\n(COVID-19 impact)',
    xy=(2020, visits_by_year.loc[visits_by_year['Year'] == 2020, 'Estimate'].values[0]),
    xytext=(2017.8, visits_by_year['Estimate'].max() * 0.85),
    arrowprops=dict(arrowstyle='->', color='#972eb9'),
    fontsize=11,
    color='#972eb9'
)

plt.legend(frameon=False)

plt.tight_layout()
plt.show()
```

This graph shows the total number of Emergency Department (ED) visits in the United States from 2016 to 2020. From 2016 through 2019, ED visits remain relatively stable with a slight upward trend, indicating consistent use of emergency care across these years.

In 2020, there is a sharp and notable decline in ED visits, which is clearly highlighted in the visualization. This drop aligns with the onset of the COVID-19 pandemic, when lockdowns, fear of exposure, and changes in healthcare-seeking behavior led many individuals to delay or avoid emergency care.

Overall, the graph effectively tells a clear story: ED utilization was steady before 2020, followed by a dramatic disruption during the pandemic, emphasizing how external public health crises can significantly alter healthcare access and behavior.

```python
## Folium Map of ED Visites rates by U.S Regions (2020)

import pandas as pd
import folium
year = 2020
map_df = region_trend[region_trend["Year"] == year].copy()
map_df['Estimate'] = pd.to_numeric(map_df['Estimate'], errors='coerce')
map_df.dropna(subset=['Estimate'], inplace=True)

region_coords = {
    "Northeast": [42.5, -75.5],
    "Midwest": [41.8, -87.5],
    "South": [33.0, -84.0],
    "West": [39.0, -120.0]
}
region_colors = {
    "Northeast": "red",
    "Midwest": "green",
    "South": "orange",
    "West": "purple"
}
m = folium.Map(location=[39, -98], zoom_start=4)

for _, row in map_df.iterrows():
    region_name = row["Subgroup"]
    if region_name in region_coords:
        lat, lon = region_coords[region_name]
        folium.CircleMarker(
            location=[lat, lon],
            radius=row["Estimate"] / 10,
            popup=f"{region_name}: {row['Estimate']:.2f} visits per 1,000",
            fill=True,
            fill_opacity=0.7,
            color=region_colors.get(region_name, "blue"),
            fill_color=region_colors.get(region_name, "blue")
        ).add_to(m)

title_html = f'''
     <div style="position: fixed;
                 top: 10px; left: 50px; width: 500px; height: 50px;
                 z-index:9999; font-size:20px; font-weight:bold;
                 background-color: white; opacity: 0.8;
                 padding: 5px; border-radius: 10px;">
         ED Visit Rates by U.S. Region ({year})
     </div>
     '''
m.get_root().html.add_child(folium.Element(title_html))

legend_html = '''
     <div style="position: fixed;
                 bottom: 50px; left: 50px; width: 150px; height: 120px;
                 z-index:9999; font-size:14px; font-weight:bold;
                 background-color: white; opacity: 0.9;
                 padding: 10px; border-radius: 10px;">
         <b>Region Legend</b><br>
         <i style="background:red; width:10px; height:10px; display:inline-block"></i> Northeast<br>
         <i style="background:green; width:10px; height:10px; display:inline-block"></i> Midwest<br>
         <i style="background:orange; width:10px; height:10px; display:inline-block"></i> South<br>
         <i style="background:purple; width:10px; height:10px; display:inline-block"></i> West
     </div>
     '''
m.get_root().html.add_child(folium.Element(legend_html))

# Display map
m
```
